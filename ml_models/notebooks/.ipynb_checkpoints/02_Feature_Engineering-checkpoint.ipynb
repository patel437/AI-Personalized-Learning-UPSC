```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('D:/Project/AI-Personalized-Learning-UPSC/backend/data/raw/upsc_prelims_data.csv')
df.head()
```

```python
df.info()
```

### Define Subject Columns

```python
# Define GS subjects (General Studies Paper 1)
gs_subjects = ['gs_history', 'gs_geography', 'gs_polity', 'gs_economy', 
               'gs_sciencetech', 'gs_environment', 'gs_currentaffairs', 'gs_artculture']

# Define CSAT subjects (Paper 2)
csat_subjects = ['csat_comprehension', 'csat_logicalreasoning', 'csat_quantitative', 
                 'csat_datainterpretation', 'csat_decisionmaking']

# Check which subjects exist in dataset
existing_gs = [col for col in gs_subjects if col in df.columns]
existing_csat = [col for col in csat_subjects if col in df.columns]

print(f"GS Subjects found: {len(existing_gs)}")
print(f"GS Subjects: {existing_gs}")
print(f"\nCSAT Subjects found: {len(existing_csat)}")
print(f"CSAT Subjects: {existing_csat}")
```

### Convert Scores to Percentages

```python
# Convert GS subject scores to percentage (assuming each subject is out of 25)
print("Converting GS scores to percentages...")
for subject in existing_gs:
    df[f'{subject}_pct'] = df[subject] * 4
    print(f"  {subject}: converted to percentage")

# Convert CSAT scores to percentage (using min-max scaling based on max value)
print("\nConverting CSAT scores to percentages...")
for subject in existing_csat:
    max_score = df[subject].max()
    df[f'{subject}_pct'] = (df[subject] / max_score) * 100
    print(f"  {subject}: max value {max_score:.1f}, converted to percentage")

print("\n✅ Score conversion complete!")
```

```python
df.head()
```

### Create Composite Scores

```python
# Create GS Paper 1 total score (average of all GS subjects)
gs_pct_cols = [f'{sub}_pct' for sub in existing_gs]
df['gs_average_pct'] = df[gs_pct_cols].mean(axis=1)

# Create CSAT Paper 2 total score (average of all CSAT subjects)
csat_pct_cols = [f'{sub}_pct' for sub in existing_csat]
df['csat_average_pct'] = df[csat_pct_cols].mean(axis=1)

# Create overall score (70% weight to GS, 30% to CSAT as per UPSC)
df['overall_score'] = (df['gs_average_pct'] * 0.7) + (df['csat_average_pct'] * 0.3)

print("Composite scores created:")
print(f"  GS Average: {df['gs_average_pct'].mean():.1f}%")
print(f"  CSAT Average: {df['csat_average_pct'].mean():.1f}%")
print(f"  Overall Score: {df['overall_score'].mean():.1f}%")
```

### Create Performance Categories

```python
# Create performance categories based on overall score
def categorize_performance(score):
    if score >= 75:
        return 'Excellent'
    elif score >= 60:
        return 'Good'
    elif score >= 45:
        return 'Average'
    else:
        return 'Needs Improvement'

df['performance_category'] = df['overall_score'].apply(categorize_performance)

# Display distribution
print("Performance Category Distribution:")
print(df['performance_category'].value_counts())
print("\nPercentages:")
print(df['performance_category'].value_counts(normalize=True) * 100)
```

### Identify Weak and Strong Areas

```python
# Count weak subjects (score below 50%)
df['weak_subjects_count'] = (df[gs_pct_cols] < 50).sum(axis=1)

# Count very weak subjects (score below 35%)
df['very_weak_subjects'] = (df[gs_pct_cols] < 35).sum(axis=1)

# Count strong subjects (score above 75%)
df['strong_subjects_count'] = (df[gs_pct_cols] > 75).sum(axis=1)

# Find weakest subject for each student
df['weakest_subject'] = df[gs_pct_cols].idxmin(axis=1)
df['weakest_subject'] = df['weakest_subject'].str.replace('_pct', '').str.replace('gs_', '')

# Find weakest subject score
df['weakest_subject_score'] = df[gs_pct_cols].min(axis=1)

# Find strongest subject
df['strongest_subject'] = df[gs_pct_cols].idxmax(axis=1)
df['strongest_subject'] = df['strongest_subject'].str.replace('_pct', '').str.replace('gs_', '')

# Find strongest subject score
df['strongest_subject_score'] = df[gs_pct_cols].max(axis=1)

print("Weak/Strong areas identified:")
print(f"  Average weak subjects per student: {df['weak_subjects_count'].mean():.1f}")
print(f"  Average strong subjects per student: {df['strong_subjects_count'].mean():.1f}")
```

### Calculate Study Efficiency Metrics

```python
# Study efficiency (score per study hour)
df['study_efficiency'] = df['overall_score'] / (df['daily_study_hours'] + 1)

# Mock test efficiency (score per mock test)
df['mock_efficiency'] = df['overall_score'] / (df['mock_tests_attempted'] + 1)

# Calculate consistency (low variance means consistent performance)
df['subject_variance'] = df[gs_pct_cols].var(axis=1)
max_variance = df['subject_variance'].max()
df['consistency_score'] = 100 - (df['subject_variance'] / max_variance * 100)

print("Efficiency metrics calculated:")
print(f"  Average study efficiency: {df['study_efficiency'].mean():.2f}")
print(f"  Average consistency score: {df['consistency_score'].mean():.1f}%")
```

### Calculate Improvement Potential

```python
# Improvement needed to reach 70% (good qualifying score)
df['improvement_needed'] = 70 - df['overall_score']
df['improvement_needed'] = df['improvement_needed'].clip(lower=0)

# Quick win opportunities (subjects between 40-60% that can improve fast)
quick_win_mask = (df[gs_pct_cols] >= 40) & (df[gs_pct_cols] < 60)
df['quick_wins_count'] = quick_win_mask.sum(axis=1)

# Estimated hours needed to reach target (assuming 1% improvement needs 2 study hours)
df['estimated_hours_needed'] = df['improvement_needed'] * 2

print("Improvement metrics calculated:")
print(f"  Students needing improvement: {(df['improvement_needed'] > 0).sum()}")
print(f"  Average improvement needed: {df['improvement_needed'].mean():.1f}%")
print(f"  Average quick wins per student: {df['quick_wins_count'].mean():.1f}")
```

### Create Engagement Features

```python
# Test readiness score (average of quiz score and test attendance)
if 'test_attendance_percent' in df.columns:
    df['test_readiness'] = (df['test_attendance_percent'] + df['avg_quiz_score']) / 2
else:
    df['test_readiness'] = df['avg_quiz_score']

# Study streak quality (streak days × daily study hours)
if 'study_streak_days' in df.columns:
    df['streak_quality'] = df['study_streak_days'] * df['daily_study_hours']
else:
    df['streak_quality'] = df['daily_study_hours'] * 30  # default assumption

# Revision effectiveness
df['revision_effectiveness'] = df['revision_cycles'] * (df['avg_quiz_score'] / 100)

# Weekly total study effort
df['weekly_study_hours'] = df['daily_study_hours'] * 7

# Preparation intensity (months × daily hours)
df['preparation_intensity'] = df['prep_months'] * df['daily_study_hours']

print("Engagement features created:")
print(f"  Average test readiness: {df['test_readiness'].mean():.1f}")
print(f"  Average revision effectiveness: {df['revision_effectiveness'].mean():.2f}")
```

### Create Risk Scores

```python
# CSAT risk (CSAT is qualifying in nature, need 33% i.e., 66 marks out of 200)
# Converting to our percentage scale (need ~33% in CSAT)
df['csat_risk'] = df['csat_average_pct'].apply(lambda x: 'High' if x < 33 else ('Medium' if x < 50 else 'Low'))

# Overall risk based on overall score
def get_risk_level(score):
    if score < 40:
        return 'High'
    elif score < 55:
        return 'Medium'
    else:
        return 'Low'

df['overall_risk'] = df['overall_score'].apply(get_risk_level)

# Qualification probability (simple logistic function)
df['qualification_probability'] = 1 / (1 + np.exp(-(df['overall_score'] - 50) / 15))

print("Risk scores created:")
print("\nCSAT Risk Distribution:")
print(df['csat_risk'].value_counts())
print("\nOverall Risk Distribution:")
print(df['overall_risk'].value_counts())
```

### Encode Categorical Variables

```python
# List of categorical columns to encode
categorical_columns = []

if 'gender' in df.columns:
    categorical_columns.append('gender')
if 'coaching' in df.columns:
    categorical_columns.append('coaching')
if 'graduation_stream' in df.columns:
    categorical_columns.append('graduation_stream')

print(f"Encoding categorical columns: {categorical_columns}")

# Apply label encoding
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[f'{col}_encoded'] = le.fit_transform(df[col])
    label_encoders[col] = le
    print(f"  {col}: encoded to {df[f'{col}_encoded'].nunique()} unique values")

# Also encode target if exists
if 'target' in df.columns:
    target_le = LabelEncoder()
    df['target_encoded'] = target_le.fit_transform(df['target'])
    label_encoders['target'] = target_le
    print(f"\n  Target variable encoded")
```

### Create Binary Features

```python
# Create binary flags for quick filtering
df['has_coaching'] = df['coaching'].apply(lambda x: 1 if x != 'Self Study' else 0) if 'coaching' in df.columns else 0

df['regular_quiz_taker'] = (df['weekly_quizzes'] >= 4).astype(int)

df['high_mock_tester'] = (df['mock_tests_attempted'] >= 50).astype(int)

df['consistent_student'] = (df['study_streak_days'] >= 30).astype(int) if 'study_streak_days' in df.columns else 0

df['ncert_reader'] = df['ncert_read'] if 'ncert_read' in df.columns else 0

df['standard_books_user'] = df['standard_books'] if 'standard_books' in df.columns else 0

print("Binary features created:")
print(f"  Students with coaching: {df['has_coaching'].sum()}")
print(f"  Regular quiz takers: {df['regular_quiz_taker'].sum()}")
print(f"  High mock testers: {df['high_mock_tester'].sum()}")
```

### Select Features for Modeling

```python
# Define features to use for ML model
feature_columns = [
    # Demographics
    'age', 'prep_months', 'daily_study_hours', 'attempt_number',
    
    # Performance metrics
    'gs_average_pct', 'csat_average_pct', 'overall_score',
    
    # Weak/strong indicators
    'weak_subjects_count', 'very_weak_subjects', 'strong_subjects_count',
    'weakest_subject_score', 'strongest_subject_score',
    
    # Efficiency metrics
    'study_efficiency', 'mock_efficiency', 'consistency_score',
    
    # Improvement metrics
    'improvement_needed', 'quick_wins_count', 'estimated_hours_needed',
    
    # Engagement
    'mock_tests_attempted', 'weekly_quizzes', 'avg_quiz_score',
    'test_readiness', 'revision_effectiveness', 'weekly_study_hours',
    'preparation_intensity',
    
    # Binary flags
    'has_coaching', 'regular_quiz_taker', 'high_mock_tester',
    'ncert_reader', 'standard_books_user'
]

# Add encoded categorical columns if they exist
if 'gender_encoded' in df.columns:
    feature_columns.append('gender_encoded')
if 'graduation_stream_encoded' in df.columns:
    feature_columns.append('graduation_stream_encoded')

# Remove any columns that don't exist
existing_features = [col for col in feature_columns if col in df.columns]

print(f"Selected {len(existing_features)} features for modeling:")
print("\nFeature list:")
for i, feature in enumerate(existing_features, 1):
    print(f"  {i}. {feature}")
```

### Prepare Final Dataset

```python
# Create feature matrix X
X = df[existing_features].copy()

# Create target variable y (if target column exists)
if 'target' in df.columns:
    y = df['target']
    print("Target variable: 'target' (1=Qualified, 0=Not Qualified)")
elif 'target_encoded' in df.columns:
    y = df['target_encoded']
    print("Target variable: 'target_encoded'")
else:
    y = None
    print("Warning: No target column found!")

print(f"\nFeatures shape: {X.shape}")
if y is not None:
    print(f"Target shape: {y.shape}")
    print(f"\nTarget distribution:")
    print(y.value_counts())
```
