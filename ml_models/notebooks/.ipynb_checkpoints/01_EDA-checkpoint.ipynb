## VISUALIZATION

### Target Variable Analysis 

```python
# Check distribution of target variable (passed/failed)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot
target_counts = df['target'].value_counts()
axes[0].bar(['Not Qualified (0)', 'Qualified (1)'], target_counts.values, 
            color=['lightcoral', 'lightgreen'], edgecolor='black')
axes[0].set_title('Distribution of Target Variable', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Qualification Status')
axes[0].set_ylabel('Number of Students')
axes[0].set_ylim(0, 800)

# Add value labels on bars
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 10, str(v), ha='center', fontweight='bold')

# Pie chart
colors = ['lightcoral', 'lightgreen']
explode = (0.05, 0)
axes[1].pie(target_counts.values, labels=['Not Qualified', 'Qualified'], 
            autopct='%1.1f%%', colors=colors, explode=explode, shadow=True)
axes[1].set_title('Target Variable Percentage', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Qualified Students: {target_counts[1]} ({target_counts[1]/len(df)*100:.1f}%)")
print(f"Not Qualified Students: {target_counts[0]} ({target_counts[0]/len(df)*100:.1f}%)")
```

### Demographic Analysis

```python


#  Overall Age Distribution

sns.histplot(df['age'], bins=15, kde=True, color='skyblue', edgecolor='black')
plt.axvline(df['age'].mean(), color='red', linestyle='--', label=f'Mean: {df["age"].mean():.1f}')
plt.title('Overall Age Distribution of Aspirants', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Average age of qualified students: {df[df['target'] == 1]['age'].mean():.1f} years")
```

```python
# Age distribution analysis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age distribution overall
axes[0,0].hist(df['age'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Age Distribution of All Students', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')
axes[0,0].axvline(df['age'].mean(), color='red', linestyle='--', label=f'Mean: {df["age"].mean():.1f}')
axes[0,0].legend()

# Age by qualification
qualified = df[df['target'] == 1]['age']
not_qualified = df[df['target'] == 0]['age']
axes[0,1].hist([qualified, not_qualified], bins=12, label=['Qualified', 'Not Qualified'], 
               alpha=0.7, color=['green', 'red'], edgecolor='black')
axes[0,1].set_title('Age Distribution by Qualification', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Age')
axes[0,1].set_ylabel('Frequency')
axes[0,1].legend()

# Gender distribution
gender_counts = df['gender'].value_counts()
axes[1,0].bar(gender_counts.index, gender_counts.values, color=['#FF9999', '#66B2FF'], edgecolor='black')
axes[1,0].set_title('Gender Distribution', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Gender')
axes[1,0].set_ylabel('Count')
for i, v in enumerate(gender_counts.values):
    axes[1,0].text(i, v + 10, str(v), ha='center', fontweight='bold')

# Qualification rate by gender
gender_target = pd.crosstab(df['gender'], df['target'], normalize='index') * 100
gender_target.plot(kind='bar', ax=axes[1,1], color=['red', 'green'], edgecolor='black')
axes[1,1].set_title('Qualification Rate by Gender', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Gender')
axes[1,1].set_ylabel('Percentage (%)')
axes[1,1].legend(['Not Qualified', 'Qualified'])
axes[1,1].set_ylim(0, 100)

# Add percentage labels
for i in range(len(gender_target)):
    axes[1,1].text(i, gender_target.iloc[i, 1] - 5, f"{gender_target.iloc[i, 1]:.1f}%", 
                   ha='center', fontweight='bold', color='white')
print(f"Average age of qualified students: {qualified.mean():.1f} years")
print(f"Average age of not qualified students: {not_qualified.mean():.1f} years")

plt.tight_layout()
plt.show()
```

### Graduation Stream Analysis

```python
# Graduation stream analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Stream distribution
stream_counts = df['graduation_stream'].value_counts()
axes[0].bar(stream_counts.index, stream_counts.values, color='lightblue', edgecolor='black')
axes[0].set_title('Distribution by Graduation Stream', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Stream')
axes[0].set_ylabel('Number of Students')
axes[0].tick_params(axis='x', rotation=45)

for i, v in enumerate(stream_counts.values):
    axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold')

# Qualification rate by stream
stream_target = pd.crosstab(df['graduation_stream'], df['target'], normalize='index') * 100
stream_target_sorted = stream_target.sort_values(by=1, ascending=False)
stream_target_sorted[1].plot(kind='bar', ax=axes[1], color='green', edgecolor='black', alpha=0.7)
axes[1].set_title('Qualification Rate by Graduation Stream', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Stream')
axes[1].set_ylabel('Qualification Rate (%)')
axes[1].tick_params(axis='x', rotation=45)
axes[1].set_ylim(0, 50)

# Add percentage labels
for i, v in enumerate(stream_target_sorted[1].values):
    axes[1].text(i, v + 1, f"{v:.1f}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nQualification rate by stream:")
print(stream_target[1].sort_values(ascending=False))
```

###  Study Resources Analysis

```python
# Analysis of study resources used
resource_cols = ['ncert_read', 'standard_books', 'coaching', 'online_resources']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, col in enumerate(resource_cols):
    row = idx // 2
    col_pos = idx % 2
    
    # Count of students using each resource
    resource_counts = df[col].value_counts()
    
    # Qualification rate for those who used vs didn't use
    used_qualified = df[df[col] == 1]['target'].mean() * 100
    not_used_qualified = df[df[col] == 0]['target'].mean() * 100
    
    axes[row, col_pos].bar(['Used', 'Not Used'], [used_qualified, not_used_qualified], 
                           color=['green', 'red'], edgecolor='black')
    axes[row, col_pos].set_title(f'Qualification Rate - {col.replace("_", " ").title()}', 
                                 fontsize=11, fontweight='bold')
    axes[row, col_pos].set_ylabel('Qualification Rate (%)')
    axes[row, col_pos].set_ylim(0, 40)
    
    # Add percentage labels
    axes[row, col_pos].text(0, used_qualified + 1, f"{used_qualified:.1f}%", 
                            ha='center', fontweight='bold')
    axes[row, col_pos].text(1, not_used_qualified + 1, f"{not_used_qualified:.1f}%", 
                            ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Additional insight: Number of resources used
df['resources_used'] = df[resource_cols].sum(axis=1)
resources_count_qualified = df[df['target'] == 1]['resources_used'].mean()
resources_count_not_qualified = df[df['target'] == 0]['resources_used'].mean()

print(f"Average resources used - Qualified: {resources_count_qualified:.1f} out of 4")
print(f"Average resources used - Not Qualified: {resources_count_not_qualified:.1f} out of 4")
```

###  GS Scores Analysis

```python
# GS subjects analysis
gs_subjects = ['gs_history', 'gs_geography', 'gs_polity', 'gs_economy', 
               'gs_sciencetech', 'gs_environment', 'gs_currentaffairs', 'gs_artculture']

fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for idx, subject in enumerate(gs_subjects):
    # Boxplot comparing qualified vs not qualified
    data_to_plot = [df[df['target'] == 0][subject], df[df['target'] == 1][subject]]
    bp = axes[idx].boxplot(data_to_plot, labels=['Not Qual', 'Qual'], patch_artist=True,
                           boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red', linewidth=2))
    axes[idx].set_title(subject.replace('gs_', '').title(), fontsize=10, fontweight='bold')
    axes[idx].set_ylabel('Score')

plt.suptitle('GS Subject Scores Comparison (Qualified vs Not Qualified)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

```

```python
# Average scores comparison
print("\n=== Average GS Subject Scores ===")
print(f"{'Subject':<20} {'Not Qualified':<15} {'Qualified':<15} {'Difference':<10}")
print("-" * 60)
for subject in gs_subjects:
    not_qual_avg = df[df['target'] == 0][subject].mean()
    qual_avg = df[df['target'] == 1][subject].mean()
    diff = qual_avg - not_qual_avg
    print(f"{subject.replace('gs_', '').title():<20} {not_qual_avg:<15.1f} {qual_avg:<15.1f} {diff:<10.1f}")
```

```python
# Total GS score analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution of total GS score
axes[0].hist([df[df['target'] == 0]['gs_total_score'], df[df['target'] == 1]['gs_total_score']], 
             bins=20, label=['Not Qualified', 'Qualified'], alpha=0.7, color=['red', 'green'], edgecolor='black')
axes[0].set_title('Distribution of GS Total Score', fontsize=12, fontweight='bold')
axes[0].set_xlabel('GS Total Score')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Boxplot comparison
axes[1].boxplot([df[df['target'] == 0]['gs_total_score'], df[df['target'] == 1]['gs_total_score']], 
                labels=['Not Qualified', 'Qualified'], patch_artist=True,
                boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red', linewidth=2))
axes[1].set_title('GS Total Score Comparison', fontsize=12, fontweight='bold')
axes[1].set_ylabel('GS Total Score')

plt.tight_layout()
plt.show()

print(f"\nAverage GS Total Score - Qualified: {df[df['target'] == 1]['gs_total_score'].mean():.1f}")
print(f"Average GS Total Score - Not Qualified: {df[df['target'] == 0]['gs_total_score'].mean():.1f}")
```

### Correlation Analysis

```python
# Correlation analysis for important features
important_features = ['prep_months', 'daily_study_hours', 'mock_tests_attempted', 'revision_cycles', 
                      'attempt_number', 'gs_total_score', 'csat_total_score', 'study_streak_days', 
                      'test_attendance_percent', 'avg_quiz_score', 'target']

correlation_matrix = df[important_features].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1, cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix of Important Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
